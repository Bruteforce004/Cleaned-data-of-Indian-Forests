```python
df=df.replace(np.nan,'')
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.at[36,'State/Union Territory']='Total'
```

```python
cols=list(df)
for i in cols:
  for j in range(0,37):
    if type(df[i][j])==str:
      df[i][j]=df[i][j].replace('-','')
      df[i][j]=df[i][j].replace('*','')
      
df.at[0,'Actual Forest Cover-09']=float(df['Actual Forest Cover-09'][0])
df.at[0,'Actual Forest Cover-11']=float(df['Actual Forest Cover-11'][0])
```
