```python
def import_file():
  x=input("insert the url")
  y=input("is it csv or excel")
  if y=='csv':
    z=pd.read_csv(x)
    return z
  elif y=='excel':
    z=pd.read_excel(x)
    return z
  else:
    print("select again csv or excel !!!")
```
