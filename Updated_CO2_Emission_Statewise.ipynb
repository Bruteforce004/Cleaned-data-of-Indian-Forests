```python
z= z.dropna()
z
```

```python
z.columns=z.iloc[0]
```

```python
z.drop(labels=0, axis=0, inplace= True)
```

```python
z.rename(columns={'Year':'Estimated CO2 Emission by States(in 000 metric tons of C) years'},inplace= True)
```

```python
z.head()
```

```python
z.reset_index(drop=True, inplace= True)
```

```python
z.dtypes

```

```python
z['Estimated CO2 Emission by States(in 000 metric tons of C) years']=z['Estimated CO2 Emission by States(in 000 metric tons of C) years'].astype(int)
```
