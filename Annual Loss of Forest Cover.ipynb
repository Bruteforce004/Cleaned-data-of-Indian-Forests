We see rows after the 18th row cannot be considered for yearly loss of forest cover estimates, therefore we need to split the dataset.

# Splitting the Dataset

```python
df_1 = df.iloc[:18,:]
```

```python
df_1
```

```python
df_1.dtypes
```

```python
df_1.replace(np.nan,0)
```

```python
df_1['Year']=df_1['Year'].str.replace(',','')
```

```python
df_1['Tree_Cover']=df_1['Tree_Cover'].str.replace(',','')
```

```python
df_1['Forest_Cover']=df_1['Forest_Cover'].str.replace(',','')
```

```python
df_1['Tree_Cover']=df_1['Tree_Cover'].astype(float)
```

```python
df_1['Forest_Cover']=df_1['Forest_Cover'].astype(float)
```

```python
df_1['Year']=df_1['Year'].astype(int)
```

```python
df_1.dtypes
```

```python
df_1.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
s_imputer=SimpleImputer(missing_values=np.nan,strategy='mean',verbose=0)
```

```python
s_imputer=s_imputer.fit(df_1[0:8])
df_1[0:18]=s_imputer.transform(df[0:18])
```

```python
df_1
```

```python
df_1.to_csv('Annual_Loss_Of_Forest_cover.csv')
```
