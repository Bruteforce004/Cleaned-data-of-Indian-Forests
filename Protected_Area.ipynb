**Protected Areas 2006-2020**

**The Dataset**

Source: National Wildlife Database, Wildlife Institute of India

Protected areas are those in which human occupation or at least the exploitation of resources is limited.The dataset used gives the data on Protected Areas of India from 2000 to 2020 (As on December, 2020). It contains the following columns:

Year                                       
No. of National Parks                       
Area Under National Parks (km2)            
No. of Wild Life Sanctuaries               
Area Under Wild Life Sanctuaries (km2)     
No. of Community Reserves                  
Area Under Community Reserves (km2)        
No. of Conservation Reserves               
Area Under Conservation Reserves (km2)     
No. of Protected Areas                     
Total Area under Protected Areas (km2)     

These data are based on availability of data from Forest Department and Gazette Notification notified by the Ministries.

Community Reserves and Conservation Reserves have been established in India from 2006 onwards. Hence these values are zero


```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read in the dataset
p_area = pd.read_csv('https://github.com/sg-008/Data-of-Indian-Forests/blob/main/Protected_Areas.csv?raw=true',encoding='unicode_escape',header=2)
```

```python
df = pd.DataFrame(p_area)
df
```

```python
df.drop([0,16,17],inplace=True)
```

```python
df
```

```python
# Printing the header of the DataFrame
df.head()
```

```python
# Printing data types of DataFrame
df.dtypes
```

```python
# Printing info of DataFrame
df.info()
```

```python
df["Area Under National Parks (km2)"] = df["Area Under National Parks (km2)"].str.replace(",","").astype(float)
```

```python
df["Area Under Wild Life Sanctuaries (km2)"] = df["Area Under Wild Life Sanctuaries (km2)"].str.replace(",","").astype(float)
```

```python
df["Area Under Community Reserves (km2)"] = df["Area Under Community Reserves (km2)"].str.replace(",","").astype(float)
```

```python
df["Area Under Conservation Reserves (km2)"] = df["Area Under Conservation Reserves (km2)"].str.replace(",","").astype(float)
```

```python
df["Total Area under Protected Areas (km2)"] = df["Total Area under Protected Areas (km2)"].str.replace(",","").astype(float)
```

```python
df["Year"] = df["Year"].astype(int)
```

```python
df["No. of Community Reserves"] = df["No. of Community Reserves"].astype(float)
```

```python
df["No. of Conservation Reserves"] = df["No. of Conservation Reserves"].astype(float)
```
